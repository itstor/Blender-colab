##Render Parameter


*   If you want to render a single image, set only the start_frame leave the end_frame value to default
*   If you want to render all frame, leave start_fram and end_frame as default
*   If you want to render at specific range of frame, set the start_frame and the end_frame





```
output_path = "/content/Rendered/"
file_path = '/content/drive/MyDrive/PKM/Halo2.blend' #@param {type: "string"}
output_filename = 'Rendered' #@param {type: "string"}
render_engine = 'CYCLES' #@param ["CYCLES", "EEVEE"]
if render_engine == "EEVEE":
  render_engine = "BLENDER_RENDER"
render_type = "Single image" #@param  ["Single image", "All frame", "Specific range"]
start_frame = 26 #@param {type: "number"}
end_frame = 0 #@param {type: "number"}
render_format = "PNG" #@param ["TGA", "RAWTGA", "JPEG", "IRIS", "IRIZ", "AVIRAW", "AVIJPEG", "PNG", "BMP"]
output = output_path + output_filename
```

##Render Time ðŸŽ‰ðŸŽ‰ðŸŽ‰

```
#@title Just Start it
if render_type == "Single image":
  !blender -b $file_path -noaudio -o $output -E $render_engine -F $render_format -x 1 -f $start_frame
if render_type == "All frame":
  !blender -b $file_path -noaudio -x 1 -o $output -E $render_engine -F $render_format -a
if render_type == "Specific range":
  !blender -b $file_path -noaudio -x 1 -o $output -E $render_engine -F $render_format -s $start_frame -e $end_frame
```
